```python
# Task1
#a
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc as img
from scipy import ndimage as nimg

res = np.array([[1/16, 2/16, 1/16], [2/16, 4/16, 2/16], [1/16, 2/16, 1/16]])
imlen = img.imread('./lena.jpg', mode = 'F')
fig = plt.figure(figsize = (10, 15))
p1 = nimg.convolve(imlen, res, mode='nearest', cval = 0.0)
plt.subplot(221)
plt.imshow(imlen, cmap = 'gray')
plt.subplot(222)
plt.imshow(p1,  cmap = 'gray')
plt.show()

```
